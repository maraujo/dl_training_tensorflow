```python
from sklearn.datasets import load_sample_image
```

```python
import tensorflow as tf
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# Load images
china = load_sample_image("china.jpg") / 255
flower = load_sample_image("flower.jpg") / 255

images = np.array([china, flower])
batch_size, height, width, channels = images.shape
print("bs: {}, h: {}, w:{}, c:{}".format(batch_size, height, width, channels))
```

```python
filters1 = np.zeros(shape=(7,7,channels, 2), dtype=np.float32)
filters1[:, 3, :, 0] = 1
filters1[3, :, :, 1] = 1
filters2 = np.zeros(shape=(7,7,2, 1), dtype=np.float32)
filters2[:, 3, :, 0] = 1
```

```python
outputs = tf.nn.conv2d(images, filters1, strides=2, padding="SAME")
outputs2 = tf.nn.conv2d(outputs, filters2, strides=2, padding="SAME")
```
